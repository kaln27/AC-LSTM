```python
pca = PCA(n_components=16, whiten=False)
min_max_sca = MinMaxScaler()
min_case = train_fd0001[0].min()
max_case = train_fd0001[0].max()
inputs_seq = list()
for i in range(min_case, max_case+1):
    data = train_fd0001[train_fd0001[0]==i].drop(columns=[0, 1]).values
    data = pca.fit_transform(data)
    data = min_max_sca.fit_transform(data)
    inputs_seq.append(data)
```
